```python
import pandas as pd
import numpy as np
```

```python
# Question 1
# Create a DataFrame using the sample of dataset from the National Flower of Ethiopia (This was adapted from Iris data)

NFE_dict = {'sepal_length':[5.1,5,5.4,5.5,6.5,6.7,7.2,6.2], 
            'sepal_width':[3.5,3.6,3.9,2.3,2.8,3.3,3.2,2.8],
            'petal_length':[1.4,1.4,1.7,4,4.6,5.7,6,4.8],
            'petal_width':[0.2,0.2,0.4,1.3,1.5,2.1,1.8,1.8],
            'species':['Rose','Rose','Rose', 'Lily', 'Lily', 'Carnation', 'Carnation', 'Carnation']}
```

```python
NFE_df = pd.DataFrame(NFE_dict)
NFE_df
```

```python
# Question 2
#How many rows and columns are in the dataset?
NFE_df.shape
#Answer is 8 rows and 5 column
```

```python
# Question 3
#How many series are in the DataFrame?
NFE_df.columns
```

```python
# Question 4
#Rename species variable to flower_categories
NFE_df.rename(columns={'species': 'flower_categories'}, inplace=True)
NFE_df
```

```python
# Question 5
# Add sepal_length and sepal_width together and call the resulting series sepal_length_width
NFE_df['sepal_length_width'] = NFE_df['sepal_length'] + NFE_df['sepal_width']
NFE_df
```

```python
# Question 6
# Drop the column petal_width from the DataFrame
NFE_df.drop(columns='petal_width', inplace=True)
NFE_df
```

```python
# Question 7
#From 6, select a DataFrame where sepal_length is greater than and name it perfect_data
perfect_data=NFE_df[NFE_df['sepal_length'] > 5.5]
```
